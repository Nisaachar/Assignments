```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv('bmi.csv')
data

```

```python
x_values = data[['BMI']]
y_values = data[['Life expectancy']]

# x_values.reshape(-1, 1)

model = LinearRegression()
model.fit(x_values, y_values)
```

```python
prediction_input = [[21.07]]  # Double brackets to create a 2D array
predicted_value = model.predict(prediction_input)
print(predicted_value)
```

```python
data = pd.read_csv('data.csv')
data
```

```python
from sklearn.preprocessing import PolynomialFeatures
import numpy as np

x = data['Var_X'].values.reshape(-1, 1)
y = data['Var_Y']

```

```python
poly_feat = PolynomialFeatures(degree = 4)
X_poly = poly_feat.fit_transform(x)
```

```python
poly_model = LinearRegression(fit_intercept = False).fit(X_poly, y)
```

```python
print(poly_model.predict([[1.2]]))
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
data = pd.read_csv('decisionTree.csv', header = None)
data
```

```python
x = data.iloc[:, 0:2]
y = data.iloc[:,2]

```

```python
model = DecisionTreeClassifier()
model.fit(x, y)

```

```python
y_pred = model.predict(x)
y_pred
```

```python
acc = accuracy_score(y, y_pred)
print(acc)
```
