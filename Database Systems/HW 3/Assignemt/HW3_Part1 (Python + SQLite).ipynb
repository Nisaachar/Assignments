```python
import sqlite3

def main():
    # Prompt the user for connect information
    print("Please enter information to connect to a relational DBMS")
   
    database = input("Enter the database name (hw3.db): ")

    # Connect to the SQLite database
    print("Connecting to the database...")
    conn = sqlite3.connect(database)
    print("Connected.")

    # Create a cursor object
    cursor = conn.cursor()

    # (3) Execute a query to list the title of movies which has more than 15 actors.
    cursor.execute("SELECT m.title FROM movie m JOIN casting c ON m.id = c.movieid GROUP BY m.id, m.title HAVING COUNT(c.actorid) > 10;")
    
    
    rset = cursor.fetchall()

    print(">>Movie Title<<")
    
    # (4) // Iterate through the result
    for row in rset:
        print(row[0])  # (5) display the movie title.

    # Close the cursor object
    cursor.close()

    # Prepare a dynamic SQL query statement to list the information of actors
    # who presented in a given movie.
    # The query result lists the actor id, actor name, and casting order.
    # The query result should be sorted by the casting order.
    titleStr = input("\nPlease enter a movie title to query its cast (e.g., Aladdin): ")
    # (7) Prepare a dynamic SQL query statement
    cursor = conn.cursor()
    pstmt = cursor.execute("SELECT actor.id, actor.name, casting.ord FROM casting INNER JOIN actor ON casting.actorid = actor.id WHERE casting.movieid IN (SELECT id FROM movie WHERE title = ?) ORDER BY casting.ord", (titleStr,))
    rset = pstmt.fetchall()

    print(f">> {titleStr} Full Cast <<")
    # (10) Iterate through the result
    for row in rset:
        print(f"{row[0]}, {row[1]}, {row[2]}")  # (11) display the cast information

    # Close the connection
    conn.close()

if __name__ == "__main__":
    main()
```
